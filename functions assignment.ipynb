```python
# What are the advantages of using generators over regular functions ?
'''
- Generators offer several advantages over regular functions :-
  1. memeory efficiency : regular function that return the entire result at once , generators yield values one by one as needed , reducing memory consumption .
  2. lazy evaluation : generators compute values only when they are requested , which can improve performance ,
                       particularly in scenarios where not all values are required . 
  3. Concise code : generators often provide a cleaner and more readable way to create irterators compared to traditional loop .
  4. State management : generators maintain their state between calls , allowing you to resume execution from the last yield point .
  5. Infinite sequences : generators can be used to create infinite sequence of value , unlike regular function that need to return a complete result.                     
'''
```

```python
# Write a python function that checks if a given numbers is prime or not from 1 to 200.
def is_prime(n):
    if n < 2 or n > 200:
        return False
    for i in range (2, int(n** 0.5) +1):
        if n % i == 0:
            return False
    return True         
```

```python
print(is_prime(7))
print(is_prime(80))
print(is_prime(202))
```

```python
# Create an iterator class in python that generates the fibonacci sequence up to a specific number of terms ?
```

```python
class FibonacciIterator:
    def _init_(self, max_terms):
        self.max_terms = max_terms
        self.count = 0
        self.a = 0
        self.b = 1
    def _iter_(self):
        return self
    def _next_(self):
        if self.count >= self.max_terms:
            raise StopIteration
        if self.count == 0:
            self.count += 1
            return self.a
        elif self.count == 1:
            self.count += 1
            return self.b
        else:
            self.a, self.b = self.b, self.a + self.b
            self.count += 1
            return self.b

fibo = FibonacciIterator(10)
for num in fibo:
    print(num)
```

```python
#6. Write a generator function in Python that yields the powers of 2 up to a given exponent.
def powers_of_two(max_exponent):
    for exponent in range(max_exponent + 1):
        yield 2 ** exponent
for power in powers_of_two(5):
    print(power)
```

```python
#7. Implement a generator function that reads a file line by line and yields each line as a string.
def read_file_line_by_line(filepath):
  """Reads a file line by line and yields each line as a string.

  Args:
    filepath: The path to the file.

  Yields:
    str: Each line from the file.
  """
  try:
    with open(filepath, 'r') as file:
      for line in file:
        yield line.rstrip('\n')  # Remove trailing newline character
  except FileNotFoundError:
    print(f"Error: File not found at {filepath}")

# Example usage:
if _name_ == "_main_":
  # Create a sample file
  with open("sample.txt", "w") as f:
    f.write("This is the first line.\n")
    f.write("This is the second line.\n")
    f.write("And this is the third line.\n")

  # Use the generator function to read and print the file
  for line in read_file_line_by_line("sample.txt"):
    print(f"Line: {line}")
```

```python
#9. Write a Python program that uses map() to convert a list of temperatures from Celsius to Fahrenheit.

def celsius_to_fahrenheit(celsius):
    return (celsius * 9/5) + 32
celsius_temperatures = [0, 20, 30, 40, 100]
fahrenheit_temperatures = list(map(celsius_to_fahrenheit, celsius_temperatures))

print("Celsius temperatures:", celsius_temperatures)
print("Fahrenheit temperatures:", fahrenheit_temperatures)
```

```python
#10. Create a Python program that uses filter() to remove all the vowels from a given string.
def remove_vowels(input_string):
    # Define a function that returns True for non-vowel characters
    def is_not_vowel(char):
        return char.lower() not in 'aeiou'

    # Use filter to keep only non-vowel characters
    filtered_chars = filter(is_not_vowel, input_string)

    # Join the filtered characters back into a string
    result = ''.join(filtered_chars)

    return result

# Example usage
input_string = "Hello, World!"
output_string = remove_vowels(input_string)
print("Original String:", input_string)
print("String without vowels:", output_string)
```
